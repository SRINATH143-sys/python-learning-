### calculating Letter Grade
 - 2 or under D
 - 3 or under C
 - 3.6 under B
 - anything above 3.6 A

```python
Student_GPA = float(input("what is your GPA?"))

if Student_GPA <=2:
    Letter_Grade = "D"
elif Student_GPA <=3:
    Letter_Grade ="C"
elif Student_GPA <=3.6:
    Letter_Grade = "B"
else:
    Letter_Grade = "A"

print (f"{name}'s Letter Grade is {Letter_Grade}")

    
```
